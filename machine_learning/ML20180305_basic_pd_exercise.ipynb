```python
import pandas as pd
```

```python
df = pd.read_csv('csvs/AirPassengers.csv')
```

```python
df.head()
```

```python
df['accum'] = df['#Passengers'].cumsum()
df['max'] = df['#Passengers'].cummax()
df['min'] = df['#Passengers'].cummin()
df.head()
```

```python
temp_date = df['Month'].map(lambda x: x.split('-'))
```

```python
temp_date
```

```python
import numpy as np
```

```python
np_date = np.array(temp_date.values.tolist())
df['year'] = np_date[:,0]
df['month'] = np_date[:,1]
df.head()
```

```python
df['diff'] = df['#Passengers'].diff().fillna(0)
df.head()
```

```python
df['diff(%)'] = df['#Passengers'].pct_change().map(lambda x: x*100).map(lambda x: '%.2f' % x) # percent change
df.head()
```

```python
df.groupby(['year']).sum()
```

```python
import dateutil
```

```python
df_phone = pd.read_csv('csvs/phone_data.csv')
df_phone.head()
```

```python
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst = True)
df_phone.head()
```

```python
df_phone.groupby(['month'])['duration'].sum()
```

```python
df_phone.groupby(['month','item'])['date'].count()
```

```python
df_phone.groupby(['month','item']).agg({'duration':sum, 'item':'count'})
```

```python
df_phone.groupby(['month','item']).agg({'duration':[min,max,sum], 'item':'count'})
```
